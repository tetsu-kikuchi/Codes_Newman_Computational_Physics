# Excercise 4.3 : Calculating derivatives

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_derivative(delta, a):
    """
    returns the derivative of f(x)=x(x-1) at x=a, 
    using the formula (f(a+delta) - f(a))/delta
    """
    def f(x):
        return x*(x-1)
    print("f(a+delta) =", f(a+delta))
    print("f(a)       =", f(a))
    return (f(a+delta) - f(a))/delta
```

```python
def plot_derivatives(a):

    powers = np.arange(4,16,2)
    results = []
    for power in powers:
        print(f"--------power={power}-------")
        delta = 10.**(-power)
        derivative = get_derivative(delta, a)
        print("derivative =", derivative)
        results.append(derivative)

    # the exact form of detivative is 2*x-1
    results = np.array(results)
    plt.plot(powers, results)
    plt.plot(powers, (2*a-1)*np.ones_like(results), "black", ls="--")
    plt.show()
```

```python
plot_derivatives(1.)
plot_derivatives(3.)
plot_derivatives(-2.)
```
