```python
from vpython import sphere, rate, vector, canvas, color, mag
from math import cos,sin,pi 
import numpy as np
```

### First example: a circularly orbiting sphere

```python
s = sphere(pos=vector(1,0,0), radius=0.1) 
for theta in np.arange(0,100*pi,0.1): 
    rate(30) 
    x = cos(theta) 
    y = sin(theta) 
    s.pos = vector(x,y,0)
```

### Second example: sun and earth (with making the earth's trail)

```python
# initial condition
sun = sphere(pos=vector(0,0,0), radius=0.2, color=color.yellow)
earth = sphere(pos=vector(1,0,0), radius=0.1, color=color.blue, make_trail=True)
v = vector(0, 1, 0)  

# physical parameters
G = 1  
# numerical settings
dt = 0.01

# make animation by solving equation of motion
while True:
    rate(100)
    r = earth.pos - sun.pos
    a = -G * r / mag(r)**3
    v += a * dt
    earth.pos += v * dt
```
