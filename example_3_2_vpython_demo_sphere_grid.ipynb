## Example 3.2 : vpython demonstration 

### What you will learn

vpython is a quick and simple visualization tool.  
In this notebook, we demonstrate its use by making a 3D sphere grid.

You can change view by your mouse:
 - scroll: zoom in and out
 - drag by right click: change angle

```python
from vpython import sphere, color, vector  # the library "visual" is used in the textbook, but seems no longer supported
```

```python
# create a white sphere
sphere()
```

```python
# add a red sphere 
sphere(radius=0.5, pos=vector(1.0,-0.2,0.0), color=color.red)
```

```python
# create a grid made of spheres
L=5
R=0.3
for x in range(-L, L+1):
    for y in range(-L, L+1):
        for z in range(-L, L+1):
            sphere(radius=R, pos=vector(x,y,z), color=color.blue)
```
